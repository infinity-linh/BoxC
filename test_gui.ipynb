```python
import wave
import pyaudio
import librosa
import tkinter as tk
from tkinter import filedialog
from tkinter import ttk


class VideoPlayer:
    def __init__(self, master):
        self.master = master
        self.master.title("Video Player")
        self.master.geometry("500x300")
        self.timeline = 0
        # self.slider = tk.Scale(self.master, from_=self.timeline, to=100, orient="horizontal")
        self.slider = ttk.Progressbar(self.master, maximum=100, value=0, orient='horizontal', style='my.Vertical.TProgressbar')

        self.slider.pack()
    def play_wav(self):
        filename = filedialog.askopenfilename(initialdir="/", title="Select a File", filetypes=[("WAV files", "*.wav")])
        if filename:
            samples, sample_rate = librosa.load(filename, sr=None)
            print(len(samples))
            # Play the audio file
            distance = len(samples)//100
            p = pyaudio.PyAudio()
            stream = p.open(format=pyaudio.paFloat32,
                            channels=1,
                            rate=sample_rate,
                            output=True)
            for idx in range(100):
                self.slider.step(idx)
                stream.write(samples[idx*distance:(idx+1)*distance].tobytes())
            stream.write(samples[100*distance:].tobytes())
            # self.slider.step(idx)
            stream.stop_stream()
            stream.close()
            p.terminate()
        
root = tk.Tk()
app = VideoPlayer(root)
play_button = tk.Button(root, text="Play WAV", command=app.play_wav)
play_button.pack(pady=20)
root.mainloop()
```

```python
import customtkinter
from PIL import Image, ImageTk
import cv2

customtkinter.set_appearance_mode("System")  # Modes: "System" (standard), "Dark", "Light"
customtkinter.set_default_color_theme("dark-blue")  # Themes: "blue" (standard), "green", "dark-blue"

cap = cv2.VideoCapture(0)

class TimeIn(customtkinter.CTk):

    def __init__(self):
        super().__init__()
        
        self.title("Time In")
        
        # =================== Center Form =================== #
        window_height = 800
        window_width = 1000

        screen_width = self.winfo_screenwidth()
        screen_height = self.winfo_screenheight()

        x_cordinate = int((screen_width/2) - (window_width/2))
        y_cordinate = int((screen_height/3) - (window_height/3))
        
        self.geometry("{}x{}+{}+{}".format(window_width, window_height, x_cordinate, y_cordinate))
        
         
        
        # =================== Mainframe =================== #
        self.mainFrame = customtkinter.CTkFrame(master=self)
        self.mainFrame.pack(pady=15, padx=15, fill="both", expand=True)

        
        # =================== set row and column =================== #
        self.grid_columnconfigure(0, weight=1)
        self.grid_rowconfigure(1, weight=1)
        
        # =================== Camera Streaming =================== #
        self.cameraFrame = customtkinter.CTkFrame(master=self.mainFrame, width=950 )
        self.cameraFrame.grid(row=0, column=0, sticky="nswe", padx=10, pady=10)
        
        self.camera = customtkinter.CTkLabel(self.cameraFrame)
        self.camera.grid(row=0, column=0, sticky="nswe", padx=10, pady=10)


        # self.camera.configure(image=)
    
    # code for video streaming
    def camera(self):
        ret, img = cap.read()
        cv2image= cv2.cvtColor(cap.read()[1],cv2.COLOR_BGR2RGB)
        img = Image.fromarray(cv2image)
        ImgTks = ImageTk.PhotoImage(image=img)
        self.camera.imgtk = ImgTks
        self.camera.configure(image=ImgTks)
        self.after(20,self.camera)
        
    
if __name__ == "__main__":
    app = TimeIn()
    app.camera()
    app.mainloop()
    

```

```python
import tkinter as tk
from tkinter import ttk
from pygame import mixer
import os

class MusicPlayer:
    def __init__(self, root):
        self.root = root
        self.root.title("Music Player")
        self.root.geometry("400x200")

        self.music_file = "path_to_your_music_file.mp3"
        self.playing = False

        self.initialize_player()

    def initialize_player(self):
        mixer.init()
        mixer.music.load(self.music_file)

        self.play_pause_button = tk.Button(self.root, text="Play", command=self.play_pause_music)
        self.play_pause_button.pack(pady=10)

        self.stop_button = tk.Button(self.root, text="Stop", command=self.stop_music)
        self.stop_button.pack(pady=10)

        self.volume_label = tk.Label(self.root, text="Volume:")
        self.volume_label.pack()

        self.volume_slider = ttk.Scale(self.root, from_=0, to=100, orient="horizontal", command=self.set_volume)
        self.volume_slider.set(50)
        self.volume_slider.pack()

        self.music_slider = ttk.Scale(self.root, from_=0, to=100, orient="horizontal", length=360, command=self.set_position)
        self.music_slider.pack(pady=10)

        self.toggle_button = tk.Button(self.root, text="Toggle Sound", command=self.toggle_sound)
        self.toggle_button.pack(pady=10)

        self.update_slider()

    def play_pause_music(self):
        if self.playing:
            mixer.music.pause()
            self.play_pause_button["text"] = "Play"
        else:
            mixer.music.unpause()
            self.play_pause_button["text"] = "Pause"
        self.playing = not self.playing

    def stop_music(self):
        mixer.music.stop()
        self.play_pause_button["text"] = "Play"
        self.playing = False
        self.music_slider.set(0)

    def set_volume(self, value):
        mixer.music.set_volume(float(value) / 100)

    def set_position(self, value):
        position = int(float(value) / 100 * mixer.music.get_length())
        mixer.music.set_pos(position)

    def toggle_sound(self):
        if mixer.get_init() and self.playing:
            mixer.music.pause()
            self.play_pause_button["text"] = "Play"
            self.playing = False
        elif mixer.get_init() and not self.playing:
            mixer.music.unpause()
            self.play_pause_button["text"] = "Pause"
            self.playing = True

    def update_slider(self):
        if self.playing:
            current_position = mixer.music.get_pos() / 1000.0  # Chuyển đổi miliseconds sang seconds
            total_length = os.path.getsize(self.music_file) / (16 * mixer.music.get_width())  # 16 bits cho mỗi mẫu
            progress_percentage = (current_position / total_length) * 100
            self.music_slider.set(progress_percentage)
        self.root.after(100, self.update_slider)

if __name__ == "__main__":
    root = tk.Tk()
    app = MusicPlayer(root)
    root.mainloop()
```

```python
import tkinter as tk

def on_key_press(event):
    key = event.keysym
    print(f"Key pressed: {key}")

def on_key_release(event):
    key = event.keysym
    print(f"Key released: {key}")

root = tk.Tk()
root.title("Keyboard Events")

# Liên kết sự kiện bấm phím và thả phím
root.bind("<KeyPress>", on_key_press)
root.bind("<KeyRelease>", on_key_release)

# Một widget để chúng ta có thể tập trung và nhận sự kiện từ bàn phím
entry = tk.Entry(root, width=40)
entry.pack(pady=20)
entry.focus_set()  # Tập trung vào widget này để bắt đầu nhận sự kiện

# Đóng ứng dụng khi nhấn ESC
root.bind("<Escape>", lambda e: root.destroy())

root.mainloop()
```

```python
import tkinter
from typing import Callable

import numpy as np
import seaborn as sns
from matplotlib.backend_bases import key_press_handler
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib.figure import Figure


def init_gui(root, update_function: Callable) -> FigureCanvasTkAgg:
    def event_key_press(event):
        print("you pressed {}".format(event.key))
        update_function()
        key_press_handler(event, canvas)

    # create empty figure and draw
    init_figure = create_figure()
    canvas = FigureCanvasTkAgg(init_figure, master=root)
    canvas.draw()
    canvas.get_tk_widget().pack(side=tkinter.TOP, fill=tkinter.BOTH, expand=1)
    # call key press event
    canvas.mpl_connect("key_press_event", event_key_press)
    return canvas


def create_figure() -> Figure:
    # generate some data
    matrix = np.random.randint(20, size=(10, 10))
    # plot the data
    figure = Figure(figsize=(6, 6))
    ax = figure.subplots()
    sns.barplot(matrix, ax=ax, errorbar=None)
    return figure


def redraw_figure():
    figure = create_figure()
    canvas.figure = figure
    canvas.draw()


sns.set()
root = tkinter.Tk()
canvas = init_gui(root, update_function=redraw_figure)

tkinter.mainloop()
```

```python
# Nhập thư viện time
import time

# Hàm để in ra một chuỗi ký tự theo hiệu ứng gõ chữ
def print_typing(text, delay=0.1):
  # Duyệt qua từng ký tự trong chuỗi
  for char in text:
    # In ra ký tự đó và dùng tham số end để không xuống dòng
    print(char, end="", flush=True)
    # Dừng một khoảng thời gian nhỏ để tạo hiệu ứng
    time.sleep(delay)
  # Xuống dòng khi kết thúc chuỗi
  print()

# Gọi hàm với một chuỗi bất kỳ
print_typing("Xin chào tôi là linh")
```
