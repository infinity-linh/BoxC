```python
import wave
import pyaudio
import librosa
import tkinter as tk
from tkinter import filedialog
from tkinter import ttk


class VideoPlayer:
    def __init__(self, master):
        master = master
        master.title("Video Player")
        master.geometry("500x300")
        timeline = 0
        # slider = tk.Scale(master, from_=timeline, to=100, orient="horizontal")
        slider = ttk.Progressbar(master, maximum=100, value=0, orient='horizontal', style='my.Vertical.TProgressbar')

        slider.pack()
    def play_wav(self):
        filename = filedialog.askopenfilename(initialdir="/", title="Select a File", filetypes=[("WAV files", "*.wav")])
        if filename:
            samples, sample_rate = librosa.load(filename, sr=None)
            print(len(samples))
            # Play the audio file
            distance = len(samples)//100
            p = pyaudio.PyAudio()
            stream = p.open(format=pyaudio.paFloat32,
                            channels=1,
                            rate=sample_rate,
                            output=True)
            for idx in range(100):
                slider.step(idx)
                stream.write(samples[idx*distance:(idx+1)*distance].tobytes())
            stream.write(samples[100*distance:].tobytes())
            # slider.step(idx)
            stream.stop_stream()
            stream.close()
            p.terminate()
        
root = tk.Tk()
app = VideoPlayer(root)
play_button = tk.Button(root, text="Play WAV", command=app.play_wav)
play_button.pack(pady=20)
root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk
from pygame import mixer
import os

class MusicPlayer:
    def __init__(self, root):
        root = root
        root.title("Music Player")
        root.geometry("400x200")

        music_file = "path_to_your_music_file.mp3"
        playing = False

        initialize_player()

    def initialize_player(self):
        mixer.init()
        mixer.music.load(music_file)

        play_pause_button = tk.Button(root, text="Play", command=play_pause_music)
        play_pause_button.pack(pady=10)

        stop_button = tk.Button(root, text="Stop", command=stop_music)
        stop_button.pack(pady=10)

        volume_label = tk.Label(root, text="Volume:")
        volume_label.pack()

        volume_slider = ttk.Scale(root, from_=0, to=100, orient="horizontal", command=set_volume)
        volume_slider.set(50)
        volume_slider.pack()

        music_slider = ttk.Scale(root, from_=0, to=100, orient="horizontal", length=360, command=set_position)
        music_slider.pack(pady=10)

        toggle_button = tk.Button(root, text="Toggle Sound", command=toggle_sound)
        toggle_button.pack(pady=10)

        update_slider()

    def play_pause_music(self):
        if playing:
            mixer.music.pause()
            play_pause_button["text"] = "Play"
        else:
            mixer.music.unpause()
            play_pause_button["text"] = "Pause"
        playing = not playing

    def stop_music(self):
        mixer.music.stop()
        play_pause_button["text"] = "Play"
        playing = False
        music_slider.set(0)

    def set_volume(self, value):
        mixer.music.set_volume(float(value) / 100)

    def set_position(self, value):
        position = int(float(value) / 100 * mixer.music.get_length())
        mixer.music.set_pos(position)

    def toggle_sound(self):
        if mixer.get_init() and playing:
            mixer.music.pause()
            play_pause_button["text"] = "Play"
            playing = False
        elif mixer.get_init() and not playing:
            mixer.music.unpause()
            play_pause_button["text"] = "Pause"
            playing = True

    def update_slider(self):
        if playing:
            current_position = mixer.music.get_pos() / 1000.0  # Chuyển đổi miliseconds sang seconds
            total_length = os.path.getsize(music_file) / (16 * mixer.music.get_width())  # 16 bits cho mỗi mẫu
            progress_percentage = (current_position / total_length) * 100
            music_slider.set(progress_percentage)
        root.after(100, update_slider)

if __name__ == "__main__":
    root = tk.Tk()
    app = MusicPlayer(root)
    root.mainloop()
```

```python
import tkinter as tk

def on_key_press(event):
    key = event.keysym
    print(f"Key pressed: {key}")

def on_key_release(event):
    key = event.keysym
    print(f"Key released: {key}")

root = tk.Tk()
root.title("Keyboard Events")

# Liên kết sự kiện bấm phím và thả phím
root.bind("<KeyPress>", on_key_press)
root.bind("<KeyRelease>", on_key_release)

# Một widget để chúng ta có thể tập trung và nhận sự kiện từ bàn phím
entry = tk.Entry(root, width=40)
entry.pack(pady=20)
entry.focus_set()  # Tập trung vào widget này để bắt đầu nhận sự kiện

# Đóng ứng dụng khi nhấn ESC
root.bind("<Escape>", lambda e: root.destroy())

root.mainloop()
```

```python
import tkinter
from typing import Callable

import numpy as np
import seaborn as sns
from matplotlib.backend_bases import key_press_handler
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib.figure import Figure


def init_gui(root, update_function: Callable) -> FigureCanvasTkAgg:
    def event_key_press(event):
        print("you pressed {}".format(event.key))
        update_function()
        key_press_handler(event, canvas)

    # create empty figure and draw
    init_figure = create_figure()
    canvas = FigureCanvasTkAgg(init_figure, master=root)
    canvas.draw()
    canvas.get_tk_widget().pack(side=tkinter.TOP, fill=tkinter.BOTH, expand=1)
    # call key press event
    canvas.mpl_connect("key_press_event", event_key_press)
    return canvas


def create_figure() -> Figure:
    # generate some data
    matrix = np.random.randint(20, size=(10, 10))
    # plot the data
    figure = Figure(figsize=(6, 6))
    ax = figure.subplots()
    sns.barplot(matrix, ax=ax, errorbar=None)
    return figure


def redraw_figure():
    figure = create_figure()
    canvas.figure = figure
    canvas.draw()


sns.set()
root = tkinter.Tk()
canvas = init_gui(root, update_function=redraw_figure)

tkinter.mainloop()
```

```python
# Nhập thư viện time
import time

# Hàm để in ra một chuỗi ký tự theo hiệu ứng gõ chữ
def print_typing(text, delay=0.1):
  # Duyệt qua từng ký tự trong chuỗi
  for char in text:
    # In ra ký tự đó và dùng tham số end để không xuống dòng
    print(char, end="", flush=True)
    # Dừng một khoảng thời gian nhỏ để tạo hiệu ứng
    time.sleep(delay)
  # Xuống dòng khi kết thúc chuỗi
  print()

# Gọi hàm với một chuỗi bất kỳ
print_typing("Xin chào tôi là linh")
```
